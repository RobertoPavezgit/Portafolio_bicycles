```python

df_bicicletas_2019.dropna(how ='all', inplace=True) # NOW I USE DROPNA FOR DROP ROWS WHEN 1 ROW HAVE ALL COLUMNS NULLS OR WITHOUT INFORMATION
df_bicicletas_2019.isnull().sum() # HERE I USE INSULL FOR COUNT THE NUMBER OF NULL IN THE DATABASE AND ALSO NEST SUM FOR KNOW WHAT IS THE TOTAL OF NULLS IN THE SPECIFIC COLUMNS
```

```python
df_bicicletas_2019['day_of_week'].unique() # NOW I SEARCH DE NUMBER WHO REPRESENT THE DAYS OF THE WEEK AND FOR THIS USE FUNCTION UNIQUE
```

```python
# NOW  I TAKE THE COLUMN RIDE_LENGHT AND CHANGE THE FORMAT AND FOR THAT USE TIMEDELTA AND USE ERRORS COERCE IN CASE THE TRANSFORM FAIL

df_bicicletas_2019['ride_lenght'] = pd.to_timedelta(df_bicicletas_2019['ride_lenght'].astype(str), errors='coerce')
```

```python
# I TRANSFORM THE DAY OF WEEK IN NAMES BECAUSE IS MORE EASER TO UNDERSTAND
df_bicicletas_2019['day_of_week'] = df_bicicletas_2019['start_time'].dt.day_name()
df_bicicletas_2019.head(15)
```

```python

#IN THIS PART OF THE CODE, I CREATE A FUNCTION DEF FOR CLASIFICATE IN DIFFERENT SECTIONS THE DAY LIKE MORNING, AFTERNOON AND NIGHT
def clasificacion_horario(hora):
  if hora >= 6 and hora <= 12:
    return "Morning"

  elif hora > 12 and hora <= 19:
    return "Afternoon"

  else:
    return "Night"


hora_extraida = df_bicicletas_2019['start_time'].dt.hour


df_bicicletas_2019['time_of_day'] = hora_extraida.apply(clasificacion_horario)
```

```python
df_bicicletas_2019.head(20) # I SEE THE CHANGES IN THE DATABASE FOR CONFIMR THE USE OF THE FUNCTION
```

```python
print(df_bicicletas_2019['time_of_day'].value_counts()) # I USE COUNTS FOR TIMES OF THE DAYS AND SEE WHICH IS THE MOSE COMMON PART OF THE DAY FOR TRIP
```

```python
df_bicicletas_2019.info() # I SEE THE CHANGE OF DATA TYPE
```

```python
df_bicicletas_2019['ride_lenght'].describe() # I USE DESCRIBE FOR SEE STATISTICS FROM RIDE _ LENGHT COLUM
```

```python

# HERE I NEED TO CHANGE THE FORMAT OF RIDE_LENGHT BECAUSE I NEED THAT MEASURE IN MINUTES
print(df_bicicletas_2019['ride_lenght'].head())
print(df_bicicletas_2019['ride_lenght'].dtype)
df_bicicletas_2019['ride_minutes'] = df_bicicletas_2019['ride_lenght'].dt.total_seconds() / 60
```

```python
df_filtrado = df_bicicletas_2019[df_bicicletas_2019['ride_minutes'] <= 60] # HERE USE A RANGE AND I FILTER ALL TRIPS OVER 60 MINUTES BECAUSE IS SO FAR FROM MEAN AND I NEED SEE BOXPLOT FOR THAT
```

```python
import matplotlib.pyplot as plt #I USE MATPLOT AND SEABORN FOR GRAPHICS
import seaborn as sns
plt.figure(figsize=(12,6))

#HERE I USE BOXPLOT AND DEFINE AXIS X AND AXIS Y
sns.boxplot(
    x='day_of_week',
    y='ride_minutes',
    data=df_filtrado,

)
# HERE IS THE NORMAL TITLE FOR CHART AND THE AXIS
plt.title('Trip duration per day of week')
plt.xlabel('Day of week')
plt.ylabel('Ride duration (minutes)')
plt.xticks(rotation=45) # THIS IS FOR ROTATE AND SEE WITHOUT PROBLEMS DE AXIS X

plt.grid(True, axis='y', alpha=0.5) #THIS IS FOR LINES IN THE AXIS Y AND ALPHA FOR MANAGE THE TRANSPARENCY OF THAT LINES
plt.show()

# THIS PART IS FOR SEE A STATS OF GRAPHIC
stats = df_filtrado.groupby('day_of_week')['ride_minutes'].describe()

print("="*60)

print(stats[['mean', '50%', '25%', '75%']].round(2))
print("="*60)
```

```python
df_filtrado_2 = df_filtrado[df_filtrado['ride_minutes'] <= 20] # HERE I USE ANOTHER  RANGE FOR SEE BETTER THE DIFFERENCE BETWEEN TIME TRIPS IN DAYS IN THE WEEK
```

```python
plt.figure(figsize=(12,6)) # THE GRAPHIC LIKE BEFORE BUT THIS HAVE A NEW RANGE

sns.boxplot(
    x='day_of_week',
    y='ride_minutes',
    data=df_filtrado_2,
    palette = 'rainbow',
)

plt.title('Trip duration per day of week')
plt.xlabel('Day of week')
plt.ylabel('Ride duration (minutes)')
plt.xticks(rotation=45)

plt.grid(True, axis='y', alpha=0.5)
plt.show()

stats = df_filtrado.groupby('day_of_week')['ride_minutes'].describe()



print("="*60)

print(stats[['mean', '50%', '25%', '75%']].round(2))
print("="*60)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

# I USE THE SAME STRUCTURE FOR VIOLIN GRAPHIC , THIS IS FOR KNOW WHAT IS THE DAY AND HOUR WITH MORE TRIPS
sns.violinplot(
    x='day_of_week',
    y='ride_minutes',
    data=df_filtrado_2,
    palette='rainbow',
    inner='box'
)

plt.title('Distribution of trip duration per day of week')
plt.xlabel('Day of week')
plt.ylabel('Ride duration (minutes)')

plt.xticks(rotation=45)

plt.grid(True, axis='y', alpha=0.4)

plt.tight_layout()
plt.show()
```
